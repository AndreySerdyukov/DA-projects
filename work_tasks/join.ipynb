## Рабочая задачка на маппинг двух таблиц

```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_excel('для саши оквэд.xlsx', sheet_name = 0)
```

```python
data_2 = pd.read_excel('comps.xlsx', sheet_name = 0)
```

```python
data_1
```

```python
data_2
```

```python
data_1['ОГРН'] = data_1['ОГРН'].apply(lambda x: '{:.0f}'.format(x))
data_1['ИНН'] = data_1['ИНН'].apply(lambda x: '{:.0f}'.format(x))
```

```python
data_1['ОГРН'] = data_1['ОГРН'].astype(str)
data_2['ОГРН'] = data_2['ОГРН'].astype(str)
```

```python
data_1
```

```python
result = pd.merge(data_1, data_2[['ОГРН', 'Код ОКВЭД2 осн.', 'Наименование ОКВЭД2 осн.']], left_on='ОГРН', right_on='ОГРН', how='left')
```

```python
columns_to_drop = ['ОКВЭД2']
result = result.drop(columns_to_drop , axis = 1)
```
